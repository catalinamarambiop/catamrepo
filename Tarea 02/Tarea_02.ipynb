```python
#Problema 2

for numero in range(1, 101):
    if numero % 10 == 0:
        print(numero, "Bip Bip")
    elif numero % 5 == 0:
        print(numero, "Bip")
```

```python
#Problema 3

puntaje1 = int(input("Ingresa primer puntaje: "))
puntaje2 = int(input("Ingresa segundo puntaje: "))

suma_puntajes = puntaje1 + puntaje2


if suma_puntajes < 50:
    print("Resultado: Nivel principiante")
elif suma_puntajes >= 50 and suma_puntajes < 80:
    print("Resultado: Nivel intermedio")
else:
    print("Resultado: Nivel pro")
```

```python
#Problema 4

edad = int(input("Ingresa tu edad: "))
le_gusta_musica_str = input("¿Te gusta la música? (sí/no): ").lower()
le_gusta_bailar_str = input("¿Te gusta bailar? (sí/no): ").lower()

le_gusta_musica = le_gusta_musica_str == 'sí'
le_gusta_bailar = le_gusta_bailar_str == 'sí'

if edad >= 18 and le_gusta_musica and le_gusta_bailar:
    print("Eres una estrella de la fiesta")
elif le_gusta_musica and not le_gusta_bailar:
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
